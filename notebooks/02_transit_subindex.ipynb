# Transit Sub-Index

Compute the transit access sub-index from a city's GTFS feed: AM-peak
frequency, stop density, and service span within a 0.5-mi buffer of each tract.


```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))  # project root on path
import zipfile, pandas as pd, geopandas as gpd
from src.processing.transit_index import build_transit_index
from src.config import GTFS_DIR, TIGER_DIR
city='chicago'
```

```python
# Requires data/raw/gtfs/<city>_gtfs.zip (run fetch_gtfs with a key first)
gtfs = GTFS_DIR / f'{city}_gtfs.zip'
if gtfs.exists():
    with zipfile.ZipFile(gtfs) as zf:
        stops = pd.read_csv(zf.open('stops.txt'))
        stop_times = pd.read_csv(zf.open('stop_times.txt'))
    tracts = gpd.read_file(TIGER_DIR / f'{city}_tracts.geojson')
    transit = build_transit_index(stops, stop_times, tracts)
    transit.describe()
else:
    print('No GTFS feed yet — transit uses a labelled placeholder in build_subindices.')
```
